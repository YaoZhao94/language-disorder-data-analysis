```python
import os

def clean(input_lines , write_file):
    i = 0
    lines = input_lines
    while i < len(lines):
        lines[i] = "*CHI: " + lines[i][2:lines[i].find('%mor')-1]
        lines[i] = lines[i].replace('<','').replace('>','')
        index = 0
        while index < len(lines[i]):
            d_1 = lines[i].find("(",index)
            a_1 = lines[i].find("[",index)
            c_1 = lines[i].find("&",index)
            c_2 = lines[i].find("+",index)
            
            if d_1 == a_1 == c_1 == c_2 == -1:
                break
            else:
                index_list = [d_1,a_1,c_1,c_2]
                index = min(x for x in index_list if x > -1)
                if lines[i][index] == "(":
                    if lines[i][index+1].isalpha():
                        d_2 = lines[i].find(")",index)
                        lines[i] = lines[i][:index]+lines[i][index+1:d_2]+lines[i][d_2+1:]                    
                        index += 1
                    else:
                        index += 1
                elif lines[i][index] == "[":
                    if lines[i][index:index+4] == "[//]" or lines[i][index:index+3] == "[/]" or lines[i][index+1] == "*":
                        index += 1
                    else:
                        a_2 = lines[i].find("]",index)
                        lines[i] = lines[i][:index] + lines[i][a_2+1:]
                        index += 1
                elif lines[i][index] == "&":
                    c_3 = max(lines[i].find(" ",index),lines[i].find(".",index))
                    lines[i] = lines[i][:index] + lines[i][c_3+1:]
                    index += 1
                elif lines[i][index] == "+":
                    c_4 = max(lines[i].find(" ",index),lines[i].find(".",index))
                    lines[i] = lines[i][:index] + lines[i][c_4+1:]
                    index += 1
                else:
                    index += 1
        i += 1

    for z in lines:
        write_file.write(z + '\n')
    write_file.close()

if not os.path.exists("SLI-Cleaned"):
    os.mkdir("SLI-Cleaned")
if not os.path.exists("TD-Cleaned"):
    os.mkdir("TD-Cleaned")
i = 1
while i < 11:
    SLI_content = open("SLI/SLI-" + str(i) + ".txt")
    TD_content = open("TD/TD-" + str(i) + ".txt")
    SLI_write_file = open("SLI-Cleaned/SLI-" + str(i) + "-Cleaned.txt",'w')
    TD_write_file = open("TD-Cleaned/TD-" + str(i) + "-Cleaned.txt",'w')
    SLI_input_lines = SLI_content.read().split('*CHI')
    TD_input_lines = TD_content.read().split('*CHI')
    SLI_input_lines.pop(0)
    TD_input_lines.pop(0)
    clean(SLI_input_lines , SLI_write_file)
    clean(TD_input_lines , TD_write_file)
    i += 1
```
