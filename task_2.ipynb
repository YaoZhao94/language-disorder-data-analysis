```python
import os
class transcript_analyse:
    
    def __init__(self, file_name):
        self.length_of_transcript = 0
        self.dict = {}
        self.repetition = 0
        self.retracting = 0
        self.grammatical = 0
        self.pause = 0
        self.file_name = file_name
        self.summary = []
        
    def analyse_transcript(self, cleaned_file):

        lines = cleaned_file
        self.length_of_transcript = len(lines)
        i = 0
        total_content = ""
        while i < len(lines):
            lines[i] = lines[i][2:-3]
            i += 1
        size = [x.replace("[/]","").replace("[//]","").replace("(.)","").replace(".","").replace("\n","") for x in lines]
        n = 0
        record = []
        set_record = ()
        while n < len(size):
            for x in size[n].split(" "):
                set_record = (x,1)
                record.append(set_record)
            n += 1
        y = 0
        while y < len(record):
            if record[y][0] not in self.dict:
                self.dict[record[y][0]] = 0
        
            self.dict[record[y][0]] += record[y][1]
            y += 1
        
        g = 0

        while g < len(lines):
            index = 0
            while index < len(lines[g]):
        
                repetition_position = lines[g].find("[/]",index)
                retracing_position = lines[g].find("[//]",index)
                grammatical_position = lines[g].find("[*",index)
                pause_position = lines[g].find("(.",index)
                if repetition_position == retracing_position == grammatical_position == pause_position == -1:
                    break
                else:
                    index_list = [repetition_position , retracing_position , grammatical_position , pause_position]
            
                    index = min(x for x in index_list if x > -1)
                    if lines[g][index:index+3] == "[/]":
                        self.repetition += 1
                        index += 1
                    elif lines[g][index:index+4] == "[//]":
                        self.retracting += 1
                        index += 1
                    elif lines[g][index:index+2] == "[*":
                        self.grammatical += 1
                        index += 1
                    elif lines[g][index:index+2] == "(.":
                        self.pause += 1
                        index += 1              
            g += 1
        
    def __str__(self,write_file):
        self.summary.append("The statistic of file " + self.file_name + " is:")
        self.summary.append("Length of the transcript: " + str(self.length_of_transcript))
        self.summary.append("Number of repetition: " + str(self.repetition))
        self.summary.append("Number of retracting: " + str(self.retracting))
        self.summary.append("Number of grammatical errors: " + str(self.grammatical))
        self.summary.append("Number of pauses: " + str(self.pause))
        self.summary.append("Size of vocabulary: ")
        self.summary.append(str(self.dict))
        self.summary.append("")
        self.summary.append("")
        
        for z in self.summary:
            write_file.write(z + "\n")
            
            
a = 1
SLI_summary_content = []
TD_summary_content = []
if not os.path.exists("statistic"):
    os.mkdir("statistic")
SLI_write_file = open("statistic/SLI_statistic.txt",'w')
TD_write_file = open("statistic/TD_statistic.txt",'w')

while a < 11:
    SLI_file_name = "SLI-Cleaned/SLI-" + str(a) + "-Cleaned.txt"
    SLI_content = open("SLI-Cleaned/SLI-" + str(a) + "-Cleaned.txt")
    SLI_input_lines = SLI_content.read()
    SLI_input_lines = SLI_input_lines.split("*CHI")
    SLI_input_lines.pop(0)
    for c in SLI_input_lines:
        SLI_summary_content.append(c)
    SLI_script_analyse = transcript_analyse(SLI_file_name)
    SLI_script_analyse.analyse_transcript(SLI_input_lines)
    SLI_script_analyse.__str__(SLI_write_file)


    
    TD_file_name = "TD-Cleaned/TD-" + str(a) + "-Cleaned.txt"
    TD_content = open("TD-Cleaned/TD-" + str(a) + "-Cleaned.txt")
    TD_input_lines = TD_content.read()
    TD_input_lines = TD_input_lines.split("*CHI")
    TD_input_lines.pop(0)
    for d in TD_input_lines:
        TD_summary_content.append(d)
    TD_script_analyse = transcript_analyse(TD_file_name)
    TD_script_analyse.analyse_transcript(TD_input_lines)
    TD_script_analyse.__str__(TD_write_file)


    a += 1
    

SLI_group_analyse = transcript_analyse("SLI_group_analyse")
SLI_group_analyse.analyse_transcript(SLI_summary_content)
SLI_group_analyse.__str__(SLI_write_file)

TD_group_analyse = transcript_analyse("TD_group_analyse")
TD_group_analyse.analyse_transcript(TD_summary_content)
TD_group_analyse.__str__(TD_write_file)

SLI_write_file.close()
TD_write_file.close()
```
